```python
#1. Import the libraries:
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import random
from PIL import Image
from tqdm import tqdm
import os
```

```python
#2. Define image properties:
img_width=128
img_height=128
img_size=(img_width,img_height)
img_channels=3
```

```python
#3. Prepare dataset for training model:
filepath_train_nevus="./train/nevus/"
filepath_train_others="./train/others/"

data_train_nevus=os.listdir(filepath_train_nevus)
data_train_others=os.listdir(filepath_train_others)

categories=[]
images=[]
for f_name in tqdm(data_train_nevus, desc="Processing images (nevus)"):
    image_resize= Image.open(filepath_train_nevus+f_name).resize(img_size)
    images.append(image_resize)
    categories.append(0)

for f_name in tqdm(data_train_others, desc="Processing images (others)"):
    image_resize= Image.open(filepath_train_others+f_name).resize(img_size)
    images.append(image_resize)
    categories.append(1)
        
df=pd.DataFrame({
    'filename':data_train_nevus+data_train_others,
    'image':images,
    'category':categories
})
```
